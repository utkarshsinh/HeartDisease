# Distribution Plot for Numeric Columns
Now that we have the dataset, we will examine the numerical columns using the distribution plot. This will enable us to determine the minimum and maximum values, and what the overall distribution is of the whole column. As an example, let's look at the distribution of 'age' in this case. With a minimum age of 30 and a maximum age of 80, most of them are between 40 and 70.<br>
Furthermore, we can visualize the probability density of the variable by making kde = true (kernel destination estimate).

```python
sns.displot(df['Age'], kde =True)
```

# Pie Charts for Categorical Dataset
Used when we have fewer categories, as after that it becomes more challenging to analyze what's going on. In this case, we are attempting to build a pie chart based on frequency. For example, we will utilize piecharts to analyze 'sex', although the analysis could have been done through distribution plots as well. Since we are only using two bars, one for males and one for females, piecharts would be more appropriate here.<br>
Considering this, we have grouped by size, furthermore, we have written the numbers in the chart showing the percentage of each category using *autopct* punctuation where *%.1f* indicates a value going up to one decimal place.

```python
df.groupby('Sex').size()
```

```python
df.groupby('Sex').size().plot(kind ='pie', autopct = '%.1f')
```

**Analysis**<br>
In order to better understand the pie charts, let's look at another category.<br>
Having 1 means the person has the disease so there is a 55% chance that the person has the disease. We can therefore draw another conclusion as this dataset is a sample and not representative of the entire nation. Therefore, if anyone gives them a sample, there is a greater than 50% chance that they will discover they have heart disease. However, if we take a sample from a random village or town, the chances are obviously much lower because here, the people who come are aware that they may have a disease, and only then are they coming and undergoing the reading and testing. The likelihood of developing a cardiac disease is significantly lower if these tests are administered to healthy individuals or persons who are chosen at random.

```python
df.groupby('HeartDisease').size().plot(kind ='pie', autopct = '%.1f')
```


A bar graph works best if we have numbers or frequencies, but if we have fewer categories to compare, a pie chart works better.

#  Pairplot

```python
sns.pairplot(df)
```

Here we have plotted 49 pairplots to identify the correlation between various factors such as Maximum heart rate, Heart disease etc. Through these pairplots we can gain several insights into the correlation between these factors. For instance, we can see that as age increases, the Maximum heartrate decreases, i.e, they are inversely proportional. 

# JointPlot

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'hex')
```

Here, we have plotted a JointPlot in the form of a hexagon, between Maximum Heart Rate and Age.

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'reg')
```

This Jointplot contains a regression line where the model has produced the best fit regression line. This line clearly demonstrates the inversely proportional relationship between Age and Maximum Heartrate, as age increases, the Maximum heartrate decreases.

```python
sns.jointplot(x = 'HeartDisease', y = 'MaxHR', data = df, kind = 'reg')
```

From the above Jointplot, we observe that Maximum Heartrate can somewhat affect HeartDisease and vice-versa but not the extent it changes with Age.
